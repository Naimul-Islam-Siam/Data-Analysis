### Using different character / symbol as separator instead of comma (,)

```python
netflix_data = pd.read_csv(f'{data_path}/netflix_titles.csv', sep="|", index_col=0) 
# the dataset has pipe | as a separator instead of comma. That's why sep="|" is added

netflix_data.head()
```

```python
netflix_data.info()
```
